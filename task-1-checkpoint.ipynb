```python
import pandas as pd 
from sklearn import linear_model
import matplotlib.pyplot as plot
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
new_var = df.drop('income',axis = 'columns')
new_var.head(5)
```

```python
salary = df.income
salary.head(5)
```

```python
%matplotlib inline
plot.xlabel('year')
plot.ylabel('income')
plot.scatter(df.year,df.income,color='red',marker='+')
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(new_var,salary)
```

```python
%matplotlib inline
plot.xlabel('year')
plot.ylabel('income')
plot.scatter(df.year,df.income,color='red',marker='+')
plot.plot(df['year'], model.predict(df[['year']]), color='blue')
```
